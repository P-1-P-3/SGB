# Clase Usuario
Esta clase se usara para crear el usuario, tambien tendra las funcinalidades que puede ejecutar el usuario.

```python
import time

class User:
    def __init__(self, idUsers, nombre, email):
        self.idUsers = idUsers # Identificador único de un usuario
        self.name    = nombre  # Nombre del usuario
        self.email   = email   # Correo del usuario
        self.books   = []      # Lista de libros que el usuario tiene en prestamo

#--- Prestar un libro
    def lendABook(self, book):
        if book.available:
            book.changeAvailability(False)  # Cambiar disponibilidad del libro
            self.books.append(book)         # Agregar libro a la lista de libros prestados al usuario
            print(f"El libro {book.title} ha sido prestado a {self.name}")
        else:
            # Si el libro no esta disponible agregamos al usuario a la Cola - lista de espera para prestamo
            book.enqueueUser(self)
            print(f"El libro {book.title} no está disponible.")
            print(f"{self.name} ha sido agregado a la lista de espera.")

#--- Devolver un libro
    def returnABook(self, book):
        if book in self.books:              # Verificar si el libro ha sido prestado al usuario
            book.changeAvailability(True)   # Cambiar disponibilidad del libro
            self.books.remove(book)         # Eliminar libro de la lista de libros prestados al usuario
            print(f"El libro {book.title} ha sido devuelto")
            book.dequeueUser()              # Al devolver se asigna el libro al primer usuario de la cola
        else:
            print(f"El libro {book.title} no ha sido prestado a {self.name}")

#--- Buscar Libro
    @staticmethod
    def searchBook(library):
        isbn = int(input("Ingrese el ISBN del libro: "))

        startTime = time.time();

        # Buscar libro por ISBN
        book = next((book for book in library.books if book.isbn == isbn), None);

        endTime = time.time();

        # Mostrar resultado de busqueda
        if book:
            library.bookHistory.append(book) # Agregar libro encontrado en el tope de la pila
            print(f'Libro encontrado: {book.__str__()}')
        else:
            print('No se encontro ningun libro')

        print(f'Tiempo de procesamiento: {endTime - startTime} segundos')
```

# Clase Libro
Esta clase se usa para crear un libro, tambien tendra la funcionalidad que se podra ejecutar en cada instancia del libro.

```python
class Book:
    def __init__(self, isbn, title, author, editorial, year):
        self.isbn      = isbn      #  Identificador único de un libro
        self.title     = title     #  Título del libro
        self.author    = author    #  Autor del libro
        self.editorial = editorial #  Editorial del libro
        self.year      = year      #  Año de publicación del libro
        self.available = True      #  Disponibilidad del libro
        self.users     = []        #  Cola - Usuarios que esperan para un libro prestado

#--- Mostrar información del libro
    def __str__(self):
        return f"{self.isbn} - {self.title} - {self.author} - {self.editorial} - {self.year} - {'Disponible' if self.available else 'No disponible'}"

#--- Cambiar disponibilidad del libro
    def changeAvailability(self, available):
        self.available = available
        print(f"El libro {self.title} ahora está {'disponible' if self.available else 'no disponible'}")

#--- Encolar Usuario en lista de espera para un libro prestado
    def enqueueUser(self, user: User):
        self.users.append(user)

#--- Desencolar Usuario
    def dequeueUser(self):
        # Verificar si el libro tiene usuarios que quieran solicitar el prestamo
        if self.users:
            user: User = self.users.pop(0)
            user.lendABook(self) # Se presta el libro al usuario
```

# Arbol Binario Libro

```python
from prettytable import PrettyTable

class Node:
    def __init__(self, book: Book):
        self.book  = book
        self.left  = None
        self.right = None

class BookBinaryTree:
    def __init__(self):
        self.root = None
        self.field_names = ["isbn", "Titulo", "Autor", "Editorial", "Año",]

#--- Insertar libro
    def insert(self, book: Book):
        if self.root is None:
            self.root = Node(book)
        else:
            book.isbn = int(book.isbn)
            self._insert(book, self.root)

    def _insert(self, book: Book, node: Node):
        if book.isbn < node.book.isbn:
            if node.left is None:
                node.left = Node(book)
            else:
                self._insert(book, node.left)
        else:
            if node.right is None:
                node.right = Node(book)
            else:
                self._insert(book, node.right)

#--- Mostrar Libros
    def showBooks(self):
      table = PrettyTable()
      table.field_names = self.field_names
      self._inorden(self.root, table)
      print(table)

    def _inorden(self, node: Node, table: PrettyTable):
      if node:
        self._inorden(node.left, table)
        table.add_row([
            node.book.isbn,
            node.book.title,
            node.book.author,
            node.book.editorial,
            node.book.year,
        ])
        self._inorden(node.right, table)

#--- Buscar libro por isbn
    def search(self, isbn):
        if self.root is None:
            return None
        else:
          table = PrettyTable()
          table.field_names = self.field_names

          book = self._search(isbn, self.root)
          if book:
            table.add_row([
              book.isbn,
              book.title,
              book.author,
              book.editorial,
              book.year,
            ])
            print(table)
          else:
            print("Libro no encontrado")

    def _search(self, isbn, node: Node):
        if node is None:
            return None

        if node.book.isbn == int(isbn):
            return node.book
        elif int(isbn) > node.book.isbn:
            return self._search(isbn, node.right)
        else:
            return self._search(isbn, node.left)
```

# Arbol n-ario Libro

```python
from prettytable import PrettyTable

class NodeBook:
    def __init__(self, value, type):
        self.value     = value # valor del nodo (Editorial, Autor, Libro)
        self.type     = type   # tipo del nodo (editorial, autor, libro)
        self.children = []     # Lista de nodos hijos

class BookNTree:
    def __init__(self):
        self.root        = None
        self.field_names = ["isbn", "Titulo", "Autor", "Editorial", "Año"]

#--- Insertar Editorial, Autor, Libro
    def insert(self, book: Book):
        editorial = book.editorial
        author    = book.author

        if self.root is None:
            self.root = NodeBook(editorial, "editorial")

        currentNode = self.root

        # Buscar editorial
        editorialNode = self._findChildNode(currentNode, editorial, "editorial")
        if editorialNode is None: # Crea el editorial si no se encuentra.
            editorialNode = NodeBook(editorial, "editorial")
            currentNode.children.append(editorialNode)

        currentNode = editorialNode

        # Buscar autor
        authorNode = self._findChildNode(currentNode, author, "autor")
        if authorNode is None: # Crea el autor si no se encuentra.
            authorNode = NodeBook(author, "autor")
            currentNode.children.append(authorNode)

        currentNode = authorNode

        # Insertar libro
        bookNode = self._findChildNode(currentNode, book, "libro")
        if bookNode is None:
            currentNode.children.append(NodeBook(book, "libro"))

#--- Encuentra el nodo hijo
    def _findChildNode(self, node: NodeBook, value, type):
        for child in node.children:
            if child.value == value and child.type == type:
                return child
        return None

#--- Buscar libro por editorial y autor
    def search(self, editorial, autor):
        if self.root is None:
            print("No hay libros en la biblioteca")
            return

        # Buscar editorial
        editorialNode = self._findChildNode(self.root, editorial, "editorial")
        if editorialNode is None:
            print("Editorial no encontrada")
            return

        # Buscar autor
        autorNode = self._findChildNode(editorialNode, autor, "autor")
        if autorNode is None:
            print("Autor no encontrado")
            return

        # Buscar libro
        table = PrettyTable()
        table.field_names = self.field_names

        for child in autorNode.children:
            if child.type == "libro":
                table.add_row([
                  child.value.isbn,
                  child.value.title,
                  child.value.author,
                  child.value.editorial,
                  child.value.year,
                ])
        print(table)
```

```python
import sqlite3
import pandas as pd
import json
import random

class createDatabase:
  def __init__(self):
    # Cargar datos de archivos JSON
    self.libros_data = pd.read_json('https://raw.githubusercontent.com/P-1-P-3/SGB/refs/heads/main/libros_mixtos.json')
    self.users_data  = pd.read_json('https://raw.githubusercontent.com/P-1-P-3/SGB/refs/heads/main/usuarios.json')

  def createDatabase(self):
    df_libros = pd.DataFrame(self.libros_data)
    df_users  = pd.DataFrame(self.users_data)

    # Connectar base de datos 'library.db'
    conn   = sqlite3.connect('library.db')
    cursor = conn.cursor()

    # Crear tabla 'Libros'
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS books (
            isbn TEXT PRIMARY KEY,
            title TEXT,
            author TEXT,
            editorial TEXT,
            year TEXT,
            available BOOLEAN
        )
    ''')

    # Insertar datos en tabla de Libros
    df_libros.to_sql('books', conn, if_exists='replace', index=False)

    # Crear tabla 'Usuarios'
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            idUsers TEXT PRIMARY KEY,
            name TEXT,
            email TEXT
        )
    ''')

    # Insertar datos en tabla de Usuarios
    df_users.to_sql('users', conn, if_exists='replace', index=False)

    # Crear tabla de 'Lista de usuarios en espera por libro'
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS books_users_waiting (
            book_id TEXT,
            user_id TEXT,
            PRIMARY KEY (book_id, user_id),
            FOREIGN KEY (book_id) REFERENCES books (isbn),
            FOREIGN KEY (user_id) REFERENCES users (idUsers)
        )
    ''')

    # Crear tabla de 'Prestamos'
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS loans (
            prestamo_id INTEGER PRIMARY KEY AUTOINCREMENT,
            book_isbn TEXT,
            user_id TEXT,
            FOREIGN KEY (book_isbn) REFERENCES libros (isbn),
            FOREIGN KEY (user_id) REFERENCES books (idUsers)
        )
    ''')

    # Obtener todos los ISBNs y idUsers
    unavailable_books = df_libros[df_libros['available'] == False]
    unavailable_isbns = unavailable_books['isbn'].tolist()
    all_user_ids      = df_users['idUsers'].tolist()

    # Crear registros de la tabla loans
    loans_records = []
    loan_user_ids = set() # Para evitar repetir usuarios en la lista de espera
    for isbn in unavailable_isbns:
        available_users = list(set(all_user_ids) - loan_user_ids)
        if not available_users: break  # No hay más usuarios disponibles

        user_id = random.choice(available_users)
        loan_user_ids.add(user_id)
        loans_records.append({'book_isbn': isbn, 'user_id': user_id})

    df_loans = pd.DataFrame(loans_records)
    df_loans.to_sql('loans', conn, if_exists='append', index=False)

    # Generar lista de espera solo para libros no disponibles
    waiting_list_records    = []
    num_records_to_generate = 50 # generar a ultimos 50 registros
    books_with_multiple_users = random.sample(unavailable_isbns, min(10, len(unavailable_isbns)))

    for book_isbn in books_with_multiple_users:
        eligible_users = list(set(all_user_ids) - loan_user_ids)
        if len(eligible_users) < 2: break

        num_users_for_book = random.randint(2, min(5, len(eligible_users))) # 2 a 5 usuarios por libro seleccionado
        selected_users     = random.sample(eligible_users, num_users_for_book)
        for user_id in selected_users:
            waiting_list_records.append({'book_id': book_isbn, 'user_id': user_id})

    # Agregar mas registros hasta completar 50
    while len(waiting_list_records) < num_records_to_generate:
        random_book_isbn = random.choice(unavailable_isbns)
        eligible_users = list(set(all_user_ids) - loan_user_ids)
        if not eligible_users: break

        random_user_id   = random.choice(eligible_users)
        new_record       = {'book_id': random_book_isbn, 'user_id': random_user_id}
        if new_record not in waiting_list_records:
            waiting_list_records.append(new_record)

    df_waiting_list = pd.DataFrame(waiting_list_records)
    df_waiting_list.to_sql('books_users_waiting', conn, if_exists='append', index=False)

    # Confirmar cambios y cerrar conexión
    conn.commit()
    conn.close()

    print("Base de datos 'library.db' creada correctamente con las tablas 'books', 'users' y 'books_users_waiting'")

#--- Crear base de datos 'library.db'
newDB = createDatabase()
newDB.createDatabase()
```
